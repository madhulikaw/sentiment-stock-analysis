```python
import pandas as pd

stock_df = pd.read_csv(r"C:\Users\praful\Desktop\madhulika\sentiment stock analysis\data\gme_stock_clean.csv")
sentiment_df = pd.read_csv(r"C:\Users\praful\Desktop\madhulika\sentiment stock analysis\data\gme_reddit_daily_sentiment.csv")

print(f"Stock Data: {stock_df.shape[0]} rows. Columns: {list(stock_df.columns)}")
print(f"Sentiment Data: {sentiment_df.shape[0]} rows. Columns: {list(sentiment_df.columns)}")
```

```python
#Standardized column names to lowercase 'date' for a perfect match
stock_df = stock_df.rename(columns={'Date': 'date'})

#Ensured both date columns are treated as strings or datetime for exact matching
stock_df['date'] = pd.to_datetime(stock_df['date']).dt.strftime('%Y-%m-%d')
sentiment_df['date'] = pd.to_datetime(sentiment_df['date']).dt.strftime('%Y-%m-%d')

#Merging the datasets on the 'date' column
merged_df = pd.merge(stock_df, sentiment_df, on='date', how='inner')
print(f"Merged Data: {merged_df.shape[0]} rows matching perfectly by date.")
merged_df.head(5)
```

```python
#correlation matrix between prices, volume, and sentiment
correlation_matrix = merged_df[['Close', 'Volume', 'mean_sentiment', 'total_posts']].corr()
print("Pearson Correlation Matrix:")
correlation_matrix
```

```python
merged_df.to_csv(r"C:\Users\praful\Desktop\madhulika\sentiment stock analysis\data\gme_merged_analysis.csv", index=False)
print("Master analysis dataset exported successfully to your data directory!")
```
